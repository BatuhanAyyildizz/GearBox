```python
import pandas as pd

# Dosya aynı klasörde ise sadece dosya adı yeterlidir.
file_path = "Bares_T35.xlsx"
df = pd.read_excel(file_path)

# Sütun isimlerini temizleyelim (başta/sonunda boşluk varsa)
df.columns = df.columns.str.strip()

# Eğer zaman damgası "Unnamed: 0" olarak geliyorsa yeniden adlandıralım
if "Unnamed: 0" in df.columns:
    df.rename(columns={"Unnamed: 0": "Timestamp"}, inplace=True)

# Zaman damgasını datetime formatına çevirelim
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Timestamp sütununu indeks yapalım
df.set_index('Timestamp', inplace=True)

# İlk birkaç satırı kontrol edelim
print(df.head())
```

```python
df['BRS.T35_T_GBX_OIL_1'] = df['BRS.T35_T_GBX_OIL_1'].fillna(df['BRS.T35_T_GBX_OIL_1'].rolling(window=3, min_periods=1).mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['BRS.T35_T_GBX_OIL_1'], label="GBX Oil Temp 1", color="blue")
plt.plot(df.index, df['BRS.T35_T_GBX_OIL_2'], label="GBX Oil Temp 2", color="green")
plt.title("Gearbox Yağ Sıcaklıkları Zaman Serisi")
plt.xlabel("Zaman")
plt.ylabel("Sıcaklık (°C)")
plt.legend()
plt.show()
```

```python
# Örneğin, 6 periyotluk (örneğin 1 saatlik veri ise 6 saatlik ortalama) hareketli ortalama
df['GBX_OIL1_Rolling6'] = df['BRS.T35_T_GBX_OIL_1'].rolling(window=6).mean()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['BRS.T35_T_GBX_OIL_1'], alpha=0.5, label="Gerçek GBX Oil Temp 1")
plt.plot(df.index, df['GBX_OIL1_Rolling6'], color='red', label="6 Periyotluk Hareketli Ortalama")
plt.title("Gearbox Yağ Sıcaklığı ve Hareketli Ortalama")
plt.xlabel("Zaman")
plt.ylabel("Sıcaklık (°C)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Gearbox Verileri Arası Korelasyon Matrisi")
plt.show()
```

```python
import numpy as np

# Rüzgar hızı ve gerçek güç arasındaki ilişkiyi inceleyelim:
x = df['BRS.T35_Wind_speed'].values
y = df['BRS.T35_Power'].values

# Polinom (örneğin, 3. dereceden) regresyon katsayılarını bulalım
coefficients = np.polyfit(x, y, 3)
poly_model = np.poly1d(coefficients)

# Her bir rüzgar hızı için ideal güç üretimini hesaplayalım
df['Ideal_Power'] = poly_model(df['BRS.T35_Wind_speed'])
```

```python
df['Performance_Ratio'] = df['BRS.T35_Power'] / df['Ideal_Power']
```

```python
# Basit normalizasyon: min-max scaling
min_ratio = df['Performance_Ratio'].min()
max_ratio = df['Performance_Ratio'].max()
df['SoH_Index'] = (df['Performance_Ratio'] - min_ratio) / (max_ratio - min_ratio)
```
